## VW classifier

```python
argsdatastore  = 'path_to_json'  # input data: example_for_vw.json
argsmodelstore = 'path_to_temp'  # models
argslist       = ['64', '2245']  # id_segments
```

```python
import argparse
import glob
import json
import subprocess
from os.path import sep
from random import shuffle
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from typing import List, Tuple
from math import trunc
```

```python
def get_all_file_paths(some_dir: str) -> List[str]:
    return glob.glob('{}{}*.json'.format(some_dir, sep))


def print_id_from_json(file_path: str):
    with open(file_path) as f:
        js = json.load(f)
        a = js['auditory']['segments']
        b = [(i['id'], i['views']) for i in a if i['views']>0]
        print(sorted(b, key=lambda x: x[1], reverse=True))


def get_all_and_used_only_ids(file_path: str) -> Tuple[List[str], List[str]]:
    with open(file_path) as f:
        js = json.load(f)
        segments_info = js['auditory']['segments']
        used_only = []
        all_ids = []
        for i in segments_info:
            all_ids.append(i['id'])
            if i['views'] > 0:
                used_only.append(i['id'])
        return (all_ids, used_only)


def get_ids_info_from_files(folder_path: str):
    set_all_ids = set()
    set_used_only = set()
    all_files = get_all_file_paths(folder_path)

    for i in all_files:
        sai, suo = get_all_and_used_only_ids(i)
        set_all_ids.update(set(sai))
        set_used_only.update(set(suo))

    print('All docks:', len(all_files), '\n')

    print('\nall:\n')
    print(sorted(set_all_ids, key=lambda x: int(x)))

    print('\nused_only:\n')
    print(sorted(set_used_only, key=lambda x: int(x)))

    print('\nnot used:\n')
    print(sorted(set_all_ids - set_used_only, key=lambda x: int(x)))


def add_to_file(file_path: str, text: str):
    with open(file_path, 'a') as f:
        f.write(text)

def prepare_url(url: str) -> str:
    remove_list = ['://', '/', '-', '_', '.', '?', '=', '&']
    for k in remove_list:
        url = url.replace(k, " ")
    return url


def get_urls_of_segment(file_path: str, id_str: str) -> Tuple[str, str]:
    with open(file_path) as f:
        js = json.load(f)
        ids = {i['id'] for i in js['auditory']['segments'] if i['views']>0}
        url = js['rss']['channel']['item']['link']
        if id_str in ids:
            label = '1'
        else:
            label = '-1'
        return (prepare_url(url), label)


def to_vw_format(url: str, label: str = None) -> str:
    return '{} |text {}\n'.format(str(label or ''), url.lower())


def generate_sample(path_to_data: str,
                    path_to_sample: str,
                    ids_of_segments: List[str],
                    step_size: int) -> List[int]:

    all_files = get_all_file_paths(path_to_data)

    shuffle(all_files)

    test_sample_len = trunc(len(all_files) * 0.2)
    all_train_len = len(all_files) - test_sample_len
    train_sample_len_list = [i for i in range(100, all_train_len, step_size)]
    train_sample_len_list.append(all_train_len)

    # train
    for train_sample_len in train_sample_len_list:
        for train_file in all_files[:train_sample_len]:
            for segment in ids_of_segments:
                file_path = '{}{}segment_{}_train_{}.vw'.format(path_to_sample, sep, segment, train_sample_len)
                url, label = get_urls_of_segment(train_file, segment)
                add_to_file(file_path, to_vw_format(url, label))

    # test
    for test_file in all_files[-test_sample_len:]:
        for segment in ids_of_segments:
            file_path_t = '{}{}segment_{}_test.vw'.format(path_to_sample, sep, segment)
            url_t, label_t = get_urls_of_segment(test_file, segment)
            add_to_file(file_path_t, to_vw_format(url_t))
            test_labels_file = '{}{}segment_{}_test_labels.vw'.format(path_to_sample, sep, segment)
            add_to_file(test_labels_file, '{}\n'.format(label_t))

    return train_sample_len_list
```

```python
auc_res = []

train_sample_len_list = generate_sample(
    argsdatastore,
    argsmodelstore,
    argslist,
    step_size=100
)

for train_sample_len in train_sample_len_list:
    for segment in argslist:
        file_path = '{}{}segment_{}_train_{}.vw'.format(argsmodelstore, sep, segment, train_sample_len)
        model_path = '{}{}segment_{}_model_{}.vw'.format(argsmodelstore, sep, segment, train_sample_len)
        result_path = '{}{}segment_{}_predict_result_{}.txt'.format(argsmodelstore, sep, segment, train_sample_len)
        test_file_path = '{}{}segment_{}_test.vw'.format(argsmodelstore, sep, segment)
        test_labels_path = '{}{}segment_{}_test_labels.vw'.format(argsmodelstore, sep, segment)

        subprocess.check_call(['vw', '-d', file_path, '--loss_function', 'hinge', '--binary', '-f', model_path])
        subprocess.check_call(['vw', '-i', model_path, '-t', '--binary', '-d', test_file_path, '-p', result_path])

        with open(result_path) as pred_file:
            test_prediction = [float(label) for label in pred_file.readlines()]

        with open(test_labels_path) as tl_file:
            test_labels = [float(label) for label in tl_file.readlines()]

        auc_res.append((segment, train_sample_len, roc_auc_score(test_labels, test_prediction)))

print(auc_res)
```
